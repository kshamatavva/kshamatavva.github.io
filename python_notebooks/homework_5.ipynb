```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv')
```

```python
df
```

```python
melted = df.melt(
    id_vars=["Agency Name", "Square Footage"],
    value_vars=["Year Acquired", "Year Constructed"],
    var_name="Year Type",
    value_name="Year"
)

melted = melted.dropna(subset=["Square Footage", "Year"])

grouped = melted.groupby(["Agency Name", "Year Type"]).agg({"Square Footage": "sum"}).reset_index()
chart = alt.Chart(grouped).mark_bar().encode(
    x=alt.X('Agency Name:N', title='Agency', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('Square Footage:Q', title='Total Square Footage'),
    color=alt.Color('Year Type:N', title='Year Type'),
    xOffset='Year Type:N'
).properties(
    width=600,
    height=400,
    title="Total Square Footage by Agency: Year Constructed vs Year Acquired"
)
chart
```

```python
myJekyllDir = '/Users/kshamatavva/kshamatavva.github.io/assets/json/'  
chart.save(myJekyllDir + 'saved_plot4_sp25.json')
```

```python

cat = df.groupby(['County', 'Usage Description']).size().reset_index(name='Count')
usage_options = df['Usage Description'].dropna().unique().tolist()
usage_dropdown = alt.binding_select(options=usage_options, name='Type: ')
usage_selection = alt.selection_point(
    fields=['Usage Description'],
    bind=usage_dropdown,
    name='usage_param'
)

color = alt.condition(
    usage_selection,
    alt.Color('Usage Description:N', title='Usage Description', scale=alt.Scale(scheme='category10')),
    alt.value('lightgray')  
)
opacity = alt.condition(
    usage_selection,
    alt.value(1.0),
    alt.value(0.2)  
)

usage_chart = alt.Chart(cat).mark_circle(size=100).encode(
    x=alt.X('County:N', title='County', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('Count:Q', title='Number of Buildings'),
    color=color,
    opacity=opacity,
    tooltip=[
        alt.Tooltip('County:N', title='County'),
        alt.Tooltip('Usage Description:N', title='Usage'),
        alt.Tooltip('Count:Q', title='Number of Buildings')
    ]
).add_params(
    usage_selection
).properties(
    title='Building Usage by County',
    width=800,
    height=400
).interactive()
usage_chart
```

```python
myJekyllDir = '/Users/kshamatavva/kshamatavva.github.io/assets/json/'  
usage_chart.save(myJekyllDir + 'saved_plot5_sp25.json')
```
